### Discover the Male ratio per occupation and sort it from the most to the least

```python
df.groupby('occupation')['gender'].value_counts()
```

```python
def gcd(a=0, b=0): 
    if(b==0): 
        return a 
    else: 
        return gcd(b,a%b)
    
d=df.groupby('occupation')

occ=[]
mratio=[]

for k,v in d:
    m = v['gender'][v.gender=='M'].count()
    f = v['gender'][v.gender=='F'].count()
    div = gcd(m,f) 
    r = m/div
    
    occ.append(k)
    mratio.append(r)
    
    
dfr=pd.DataFrame()
dfr['Occupation'] = occ
dfr['MaleRatio']= mratio
dfr
```
