# Movie Recommendation with collaborative filtering

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv("toy_dataset.csv", index_col=0)
ratings = ratings.fillna(0)
ratings.head(5)
```

```python
def standardize(row):
    new_row = ((row-row.mean()) / row.max() -row.min())
    return new_row

ratings_std = ratings.apply(standardize)
print(ratings_std)
print("\n")
print("-"*40)

# we are transposing, since we want similarity between items which need to be in same row
item_similarity = cosine_similarity(ratings_std.T)
print("Numpy array based on which we recommend")
print(item_similarity)
```

```python
#create datafram of similarity
item_similarity_df = pd.DataFrame(item_similarity, index = ratings.columns, columns=ratings.columns)
item_similarity_df
```
